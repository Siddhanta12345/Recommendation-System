```python
import pandas as pd 
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import correlation,cosine
```

```python
data=pd.read_csv('C:\\Users\\siddh\\Downloads\\book.csv',engine='python')
data
data=data.drop(['Unnamed: 0'],axis=1)
data
```

```python
book_df=data.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
book_df.head()
```

```python
book_df.index=data['User.ID'].unique()
book_df.head()
```

```python
book_df.fillna(0,inplace=True)
book_df.head()
```

```python
books_similar=1-pairwise_distances(book_df.values,metric='cosine')
books_similar
```

```python
books_similar_df=pd.DataFrame(books_similar)
books_similar_df.head()
```

```python
books_similar_df.index=data['User.ID'].unique()
books_similar_df.columns=data['User.ID'].unique()
books_similar_df.head()
```

```python
np.fill_diagonal(books_similar,0)
books_similar_df.head()
```

```python
similar_books=book_df["You Don't Need Experience if You've Got Attitude"]
similar_books.head()
```

```python
similar_to_book=books_similar_df.corrwith(similar_books)
similar_books.head()
```
